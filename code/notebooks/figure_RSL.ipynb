```python
rate_fnames = ['RSLRate_ysv_24ka.nc', 'RSLRate_sv_26ka.nc', 'RSLRate_jdf_18ka.nc']
```

```python
### read in flowlines
jdf = pd.read_excel(f'{root}/inputs/flowlines/jdf_update.xlsx')
sv = pd.read_excel(f'{root}/inputs/flowlines/sv.xlsx')
ysv = pd.read_excel(f'{root}/inputs/flowlines/ysv_update.xlsx')
flowlines = [ysv, sv, jdf]
```

```python
### reading in RSL curves for plotting 
sv_lonlats = np.loadtxt(f'{root}/inputs/flowlines/sv_lonlats.txt')
ysv_lonlats = np.loadtxt(f'{root}/inputs/flowlines/ysv_lonlats.txt')
jdf_lonlats = np.loadtxt(f'{root}/inputs/flowlines/jdf_lonlats.txt')

lonlats = [ysv_lonlats, sv_lonlats, jdf_lonlats]

fname = 'rsl-CISLGM_icepc_022226_highres_512-prem.l50.ump2.lm3_lvz_200km.yt.mat.nc'
rsl_yt = nc.Dataset(f'{root}/inputs/rsl/{fname}')
fname = 'rsl-CISLGM_icepc_022226_highres_512-prem.l50.ump2.lm3.mx.mat.nc'
rsl_mx = nc.Dataset(f'{root}/inputs/rsl/{fname}')

rsl0_yt_plt = {}
rsl0_mx_plt = {}
inds = [2,0,2]
for i, tran in enumerate(lonlats):
    rsl0_yt_plt[i] = util.interp_time_pt(rsl_yt['rsl'][:,:,:],rsl_yt['lon'][:],rsl_yt['lat'][:],rsl_yt['t'][:],lonlats[i][inds[i],0]+360,lonlats[i][inds[i],1])
    rsl0_mx_plt[i] = util.interp_time_pt(rsl_mx['rsl'][:,:,:],rsl_mx['lon'][:],rsl_mx['lat'][:],rsl_mx['t'][:],lonlats[i][inds[i],0]+360,lonlats[i][inds[i],1])

txts = ['YSV', 'SV', 'JdF']
colors = [cs.clr_n, cs.clr_c, cs.clr_s]
```

```python
fig = pygmt.Figure()
w  = 9
h  = 3

### creating region and extents
xmin = 360 - 146.
xmax = 360 - 120.
ymin = 46.
ymax = 62.

region = [xmin,xmax,ymin,ymax]

xmins = np.array([-143, -135, -126])
xmaxs = np.array([-138, -128, -122])
ymins = np.array([58, 53, 47])
ymaxs = np.array([61, 56, 51])

projection = f'M0/0/{w}c'

res = 'full'

with pygmt.config(FONT_ANNOT_PRIMARY="8p"):
    ###########
    # delRSL Plot
    ##########
    ### Top row, rectangle
    with fig.subplot(nrows=1, ncols=1, figsize=(f"{w}c", f"{w}c")):
    
        ### plotting delRSL
        pygmt.makecpt(cmap="SCM/vik+h", series=[-200, 600])
        fig.grdimage(grid=f'{root}/outputs/delRSL_26ka17ka.nc',
                     region = region,
                     projection = projection,
                    )
    
        fig.colorbar(frame=["a200f50", "x+l@~\104@~Relative Sea Level from 26 ka to 17 ka (m)"],
                    position=Position("TC", 'outside', offset = (0,1), anchor = 'CB'),
                    orientation = 'horizontal',
                    move_text = ['annotations', 'label', 'unit']
                    )
    
        ### plotting coastlines
        fig.coast(shorelines="1/0.25p,black", region = region, projection=f'M{(xmin + xmax)/2}/{(ymin+ymax)/2}/{w}c', resolution = res)
    
        for i,flow in enumerate(flowlines):
            fig.plot(x = flow['Lon'], y = flow['Lat'], pen = f'2p,{colors[i]}',  projection=f'M{(xmin + xmax)/2}/{(ymin+ymax)/2}/{w}c', region=region)
            
            dum = np.loadtxt(f'{root}/inputs/flowlines/{transects[i]}_lonlats.txt')
            fig.plot(x = dum[0,0], y = dum[0,1],projection=f'M{(xmin + xmax)/2}/{(ymin+ymax)/2}/{w}c', region=region, style="a0.5c", fill="#FFCE1B", pen="0.5p,black")
    
            ### plotting rectangle of subplot locations
            rectangle = [[xmins[i], ymins[i], xmaxs[i], ymaxs[i]]]
            fig.plot(data=rectangle, style="r+s", pen="1p,black,-", projection=projection)
    
        ### plotting inset
        xmin_inset, xmax_inset = -180, -40
        ymin_inset, ymax_inset = 30, 70
        region_inset = [xmin_inset, xmax_inset, ymin_inset, ymax_inset]
        
        with fig.inset(
            position=Position("RT", offset=0.1),
            # box=Box(fill="white", pen="1p"),
            region=region_inset,
            projection=f"L{(-100)}/35/{ymin_inset}/{ymax_inset}/5c",
        ):
            with pygmt.config(
                MAP_FRAME_TYPE="plain",
                MAP_FRAME_PEN="1p,black",
                MAP_TICK_PEN="0p",
                FONT_ANNOT_PRIMARY="0p"):
                ### adding coastlinies
                fig.coast(shorelines="1/0.5p,black", region = region_inset, land="gray70", water="white",frame = ['wnse'])
    
            ### plotting rectangle over area
            rectangle = [[region[0], region[2], region[1], region[3]]]
            fig.plot(data=rectangle, style="r+s", pen="1p,black")
    
    
    
    ##############
    # JdF, bottom right
    ##############
    ### Move plot origin by 1 cm above the height of the entire figure
    fig.shift_origin(xshift="w+1c")
    ### right columns, shifting each at a time
    ws = 2.4
    with fig.subplot(nrows=1, ncols=1, figsize=(f"{ws}c", f"{ws}c")):
        ### setting temporary region and projection
        i = 2
        proj_dum = f'M{(xmins[i] + xmaxs[i])/2}/{(ymins[i]+ymaxs[i])/2}/{ws}c'
        reg_dum = [xmins[i], xmaxs[i], ymins[i], ymaxs[i]]
    
        ### creating colormap for rates
        pygmt.makecpt(cmap="SCM/bam+h",series=[-10, 600], truncate=(0,0.5), reverse=True, background=True)
        fig.grdimage(grid=f'{root}/outputs/{rate_fnames[i]}',
                     region = [xmins[i], xmaxs[i], ymins[i], ymaxs[i]],
                     projection = proj_dum,
                     frame = ['wSnE', 'xa2f1', 'ya2f1', ],
                    )
        
        fig.coast(shorelines="1/0.25p,black", region = reg_dum, projection=proj_dum, resolution = res, )
        fig.plot(x = flowlines[i]['Lon'], y = flowlines[i]['Lat'], pen = f'2p,{colors[i]}', projection = proj_dum, region=reg_dum)
    
    
    ##############
    # center right, SV
    ##############
    ### shift up
    fig.shift_origin(yshift="h+2c")
    with fig.subplot(nrows=1, ncols=1, figsize=(f"{ws}c", f"{ws}c")):
        i = 1
        proj_dum = f'M{(xmins[i] + xmaxs[i])/2}/{(ymins[i]+ymaxs[i])/2}/{ws}c'
        reg_dum = [xmins[i], xmaxs[i], ymins[i], ymaxs[i]]
    
        fig.grdimage(grid=f'{root}/outputs/{rate_fnames[i]}',
                     region = [xmins[i], xmaxs[i], ymins[i], ymaxs[i]],
                     projection = proj_dum,
                     frame = ['wSnE', 'a4f1']
                    )
        
        ### create colorbar
        # with pygmt.config(FONT_ANNOT_PRIMARY="12p", FONT_LABEL="16p"):
        fig.colorbar(frame=["a200f50"],
                    position=Position("RM", 'outside', offset = (1,-1), anchor = 'CM'),
                    orientation = 'vertical',
                    length = '300%'
                    )
        #"x+lRate of relative sea level change\nfollowing largest loading (mm/yr)"
        
        fig.coast(shorelines="1/0.25p,black", region = reg_dum, projection=proj_dum , resolution = res)
        fig.plot(x = flowlines[i]['Lon'], y = flowlines[i]['Lat'], pen = f'2p,{colors[i]}', projection = proj_dum, region=reg_dum)
    
    
    ##############
    # top right, SV
    ##############
    
    ### shift up
    fig.shift_origin(yshift="h-0.1c")
    with fig.subplot(nrows=1, ncols=1, figsize=(f"{ws}c", f"{ws}c")):
        i = 0
        proj_dum = f'M{(xmins[i] + xmaxs[i])/2}/{(ymins[i]+ymaxs[i])/2}/{ws}c'
        reg_dum = [xmins[i], xmaxs[i], ymins[i], ymaxs[i]]
    
        ### plotting del rsl
        fig.grdimage(grid=f'{root}/outputs/{rate_fnames[i]}',
                     region = [xmins[i], xmaxs[i], ymins[i], ymaxs[i]],
                     projection = proj_dum,
                     frame = ['wsNE', 'xa2f1', 'ya2f1']
                    )
        
        fig.coast(shorelines="1/0.25p,black", region = reg_dum, projection=proj_dum, resolution = res)
        fig.plot(x = flowlines[i]['Lon'], y = flowlines[i]['Lat'], pen = f'2p,{colors[i]}', projection = proj_dum, region=reg_dum)
    
    fig.show()
    
    
    fig.savefig(f'{root}/figures/f02_delRSL.pdf')
```
