```python
### creating scaffold of figure first
fig = pygmt.Figure()
w  = 9
h  = 3

### creating region and extents
xmin = 360 - 146.
xmax = 360 - 120.
ymin = 46.
ymax = 62.

region = [xmin,xmax,ymin,ymax]

xmins = np.array([-143, -135, -126])
xmaxs = np.array([-138, -128, -122])
ymins = np.array([58, 53, 47])
ymaxs = np.array([61, 56, 51])

projection = f'M0/0/{w}c'

res = 'full'



##############
# RLS timeseries
#############
with fig.subplot(nrows=1, ncols=1, figsize=(f"{w}c", f"{h}c")):
    fig.basemap(
        region=[15, 35, -160, 450], projection="X?", frame=["af", "WSe", "y+lRSL (m)", "x+lTime (ka)"], panel=[0, 0]
    )
    for i,tran in enumerate(transects):
        if i == 0:
            fig.plot(x=rsl_yt['t'][:], y=rsl0_yt[tran], label = f'{tran}+N3', pen = f'1p,{colors[i]}')
            fig.plot(x=rsl_yt['t'][:], y=rsl0_mx[tran], pen = f'1p,{colors[i]},-')
        else:
            fig.plot(x=rsl_yt['t'][:], y=rsl0_yt[tran], label = f'{tran}', pen = f'1p,{colors[i]}')
            fig.plot(x=rsl_yt['t'][:], y=rsl0_mx[tran], pen = f'1p,{colors[i]},-')
    fig.legend(position=Position("TR", offset=0.2), width="4.5c", box=True)

    # ### plotting lines of times of ice sheet growth
    # fig.plot(x = [35,20], y = [-140,-140], pen = f'1.5p,{clr_n}')
    # fig.plot(x = [30,20], y = [-110,-110], pen = f'1.5p,{clr_c}')
    # fig.plot(x = [20,17], y = [-90,-90], pen = f'1.5p,{clr_s}')

####################
### Times of ice growth
###################
fig.shift_origin(yshift=f"h+0.1c")
with fig.subplot(nrows=1, ncols=1, figsize=(f"{w}c", f"{1}c")):
    fig.basemap(
            region=[0, 40, 1.6e15, 2.1e15], 
            projection="X?", frame=['n'], 
            panel=[0, 0],
            )
    lw = 2.5
    ### plotting lines of times of ice sheet growth
    fig.plot(x = [35,20], y = [1.9e15,1.9e15], pen = f'{lw}p,{clr_n}')
    fig.plot(x = [30,20], y = [1.8e15,1.8e15], pen = f'{lw}p,{clr_c}')
    fig.plot(x = [35,17], y = [1.7e15,1.7e15], pen = f'{lw}p,{clr_s}')



##############
# JdF
##############
### Move plot origin by 1 cm above the height of the entire figure
fig.shift_origin(xshift="w+1c", yshift = 'h-4c')
### right columns, shifting each at a time
ws = 4.5
with fig.subplot(nrows=1, ncols=1, figsize=(f"{ws}c", f"{ws}c")):
    ### setting temporary region and projection
    i = 2
    proj_dum = f'M{(xmins[i] + xmaxs[i])/2}/{(ymins[i]+ymaxs[i])/2}/3.5c'
    reg_dum = [xmins[i], xmaxs[i], ymins[i], ymaxs[i]]

    ### creating colormap for rates
    pygmt.makecpt(cmap="SCM/bam+h", series=[-600, 600], reverse=True)
    fig.grdimage(grid=f'outputs/{rate_fnames[i]}',
                 region = [xmins[i], xmaxs[i], ymins[i], ymaxs[i]],
                 projection = proj_dum,
                 frame = ['wSnE', 'xa2f1', 'ya2f1'],
                )
    
    fig.coast(shorelines="1/0.25p,black", region = reg_dum, projection=proj_dum, resolution = res, )
    fig.plot(x = flowlines[i]['Lon'], y = flowlines[i]['Lat'], pen = '2p,white', projection = proj_dum, region=reg_dum)


##############
# center right, SV
##############
### shift up
fig.shift_origin(yshift="h+2c")
with fig.subplot(nrows=1, ncols=1, figsize=(f"{ws}c", f"{ws}c")):
    i = 1
    proj_dum = f'M{(xmins[i] + xmaxs[i])/2}/{(ymins[i]+ymaxs[i])/2}/3.5c'
    reg_dum = [xmins[i], xmaxs[i], ymins[i], ymaxs[i]]

    fig.grdimage(grid=f'outputs/{rate_fnames[i]}',
                 region = [xmins[i], xmaxs[i], ymins[i], ymaxs[i]],
                 projection = proj_dum,
                 frame = ['wSnE', 'a4f1']
                )
    
    ### create colorbar
    with pygmt.config(FONT_ANNOT_PRIMARY="12p", FONT_LABEL="16p"):
        fig.colorbar(frame=["a200f50", "x+lRate of relative sea level change (mm/yr)"],
                    position=Position("RM", 'outside', offset = (0.5,-1), anchor = 'CM'),
                    orientation = 'vertical',
                    length = '250%'
                    # move_text = ['annotations', 'label', 'unit']
                    )
    
    
    fig.coast(shorelines="1/0.25p,black", region = reg_dum, projection=proj_dum , resolution = res)
    fig.plot(x = flowlines[i]['Lon'], y = flowlines[i]['Lat'], pen = '2p,white', projection = proj_dum, region=reg_dum)


##############
# top right, SV
##############

### shift up
fig.shift_origin(yshift="h-0.75c")
with fig.subplot(nrows=1, ncols=1, figsize=(f"{ws}c", f"{ws}c")):
    i = 0
    proj_dum = f'M{(xmins[i] + xmaxs[i])/2}/{(ymins[i]+ymaxs[i])/2}/3.5c'
    reg_dum = [xmins[i], xmaxs[i], ymins[i], ymaxs[i]]

    ### plotting del rsl
    fig.grdimage(grid=f'outputs/{rate_fnames[i]}',
                 region = [xmins[i], xmaxs[i], ymins[i], ymaxs[i]],
                 projection = proj_dum,
                 frame = ['wsNE', 'xa2f1', 'ya2f1']
                )
    
    fig.coast(shorelines="1/0.25p,black", region = reg_dum, projection=proj_dum, resolution = res)
    fig.plot(x = flowlines[i]['Lon'], y = flowlines[i]['Lat'], pen = '2p,white', projection = proj_dum, region=reg_dum)

###########
# delRSL Plot
##########
fig.shift_origin(xshift="-10c", yshift = '-5.5')
### Top row, rectangle
with fig.subplot(nrows=1, ncols=1, figsize=(f"{w}c", f"{w}c")):

    ### plotting delRSL
    pygmt.makecpt(cmap="SCM/vik+h", series=[-200, 600])
    fig.grdimage(grid='outputs/delRSL_26ka17ka.nc',
                 region = region,
                 projection = projection,
                )
    with pygmt.config(FONT_ANNOT_PRIMARY="12p", FONT_LABEL="16p"):
        fig.colorbar(frame=["a200f50", "x+l@~\104@~Relative Sea Level 26 ka to 17 ka (m)"],
                    position=Position("TC", 'outside', offset = (0,1), anchor = 'CB'),
                    orientation = 'horizontal',
                    move_text = ['annotations', 'label', 'unit']
                    )

    ### plotting coastlines
    fig.coast(shorelines="1/0.25p,black", region = region, projection=f'M{(xmin + xmax)/2}/{(ymin+ymax)/2}/{w}c', resolution = res)

    for i,flow in enumerate(flowlines):
        fig.plot(x = flow['Lon'], y = flow['Lat'], pen = '2p,white',  projection=f'M{(xmin + xmax)/2}/{(ymin+ymax)/2}/{w}c', region=region)
        
        dum = np.loadtxt(f'inputs/flowlines/{transects[i]}_lonlats.txt')
        fig.plot(x = dum[0,0], y = dum[0,1],projection=f'M{(xmin + xmax)/2}/{(ymin+ymax)/2}/{w}c', region=region, style="a0.5c", fill="#FFCE1B", pen="0.5p,black")

    ### plotting inset
    xmin_inset, xmax_inset = -180, -40
    ymin_inset, ymax_inset = 30, 70
    region_inset = [xmin_inset, xmax_inset, ymin_inset, ymax_inset]
    
    with fig.inset(
        position=Position("LB", offset=0.1),
        # box=Box(fill="white", pen="1p"),
        region=region_inset,
        projection=f"L{(-100)}/35/{ymin_inset}/{ymax_inset}/5c",
    ):
        with pygmt.config(
            MAP_FRAME_TYPE="plain",
            MAP_FRAME_PEN="1p,black",
            MAP_TICK_PEN="0p",
            FONT_ANNOT_PRIMARY="0p"):
            ### adding coastlinies
            fig.coast(shorelines="1/0.5p,black", region = region_inset, land="gray70", water="white",frame = ['wnse'])

        ### plotting rectangle over area
        rectangle = [[region[0], region[2], region[1], region[3]]]
        fig.plot(data=rectangle, style="r+s", pen="1p,black")


###########
# RSL over time close up
##########
fig.shift_origin(xshift="18c", yshift = '-5c')
### Top row, rectangle
lw=1
with fig.subplot(nrows=3, ncols=1, figsize=(f"{4}c", f"{15}c"), margins = '0.5c'):
    with fig.set_panel(panel=0):  # sets the current panel
        fig.basemap(
            region=[22, 24, 0, 50],
            frame = ['wSnE', 'x+lTime (ka)', 'y+lRelative Sea Level (m)'],
            projection="X?",
        )
        fig.plot(x = rsl_yt['t'][:], y=rsl0_yt_plt[0],pen = f'{lw}p,{clr_n}')
        fig.plot(x = rsl_yt['t'][:], y=rsl0_mx_plt[0],pen = f'{lw}p,{clr_n},-')

    with fig.set_panel(panel=1):  # sets the current panel
        fig.basemap(
            region=[22, 26.3, 100, 350],
            frame = ['wSnE', 'x+lTime (ka)', 'y+lRelative Sea Level (m)'],
            projection="X?",
        )
        fig.plot(x = rsl_yt['t'][:], y=rsl0_yt_plt[1],pen = f'{lw}p,{clr_c}')
        fig.plot(x = rsl_yt['t'][:], y=rsl0_mx_plt[1],pen = f'{lw}p,{clr_c},-')
        
    with fig.set_panel(panel=2):  # sets the current panel
        fig.basemap(
            region=[16, 18.3, 0, 200],
            frame = ['nwSE', 'x+lTime (ka)', 'y+lRelative Sea Level (m)'],
            projection="X?",
        )
        fig.plot(x = rsl_yt['t'][:], y=rsl0_yt_plt[2],pen = f'{lw}p,{clr_s}')
        fig.plot(x = rsl_yt['t'][:], y=rsl0_mx_plt[2],pen = f'{lw}p,{clr_s},-')
fig.show()


# fig.savefig('figures/f02_delRSL.pdf')
```
