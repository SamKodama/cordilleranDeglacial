```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
root = str(Path.cwd().parent.parent)
import sys
sys.path.append(f'{root}/src')
import constants as cs
import os
import netCDF4 as nc
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
os.chdir('../..')
```

```python
import src.util as util
```

```python
sv_lonlats = np.loadtxt('inputs/flowlines/sv_lonlats.txt')
ysv_lonlats = np.loadtxt('inputs/flowlines/ysv_lonlats.txt')
jdf_lonlats = np.loadtxt('inputs/flowlines/jdf_lonlats.txt')
```

```python
lonlats = [ysv_lonlats, sv_lonlats, jdf_lonlats]
```

```python
fname = 'rsl-CISLGM_icepc_022226_highres_512-prem.l50.ump2.lm3_lvz_200km.yt.mat.nc'
rsl_yt = nc.Dataset(f'inputs/rsl/{fname}')
fname = 'rsl-CISLGM_icepc_022226_highres_512-prem.l50.ump2.lm3.mx.mat.nc'
rsl_mx = nc.Dataset(f'inputs/rsl/{fname}')

fname = 'rsl-CISLGM_icepc_022226_highres_512-prem.l50.ump2.lm3.yt.mat.nc'
rsl_yt_nolvz = nc.Dataset(f'inputs/rsl/{fname}')

fname = 'rsl-CISLGM_icepc_022226_highres_512-prem.l50.ump2.lm3_lvz_200km.mx.mat.nc'
rsl_mx_lvz = nc.Dataset(f'inputs/rsl/{fname}')
```

```python
rsl0_yt = {}
rsl0_mx = {}
rsl0_mx_lvz = {}
rsl0_yt_nolvz = {}
inds = [2,0,2]
for i, tran in enumerate(lonlats):
    rsl0_yt[i] = util.interp_time_pt(rsl_yt['rsl'][:,:,:],rsl_yt['lon'][:],rsl_yt['lat'][:],rsl_yt['t'][:],lonlats[i][inds[i],0]+360,lonlats[i][inds[i],1])
    rsl0_mx[i] = util.interp_time_pt(rsl_mx['rsl'][:,:,:],rsl_mx['lon'][:],rsl_mx['lat'][:],rsl_mx['t'][:],lonlats[i][inds[i],0]+360,lonlats[i][inds[i],1])
    rsl0_mx_lvz[i] = util.interp_time_pt(rsl_mx_lvz['rsl'][:,:,:],rsl_mx_lvz['lon'][:],rsl_mx_lvz['lat'][:],rsl_mx_lvz['t'][:],lonlats[i][inds[i],0]+360,lonlats[i][inds[i],1])
    rsl0_yt_nolvz[i] = util.interp_time_pt(rsl_yt_nolvz['rsl'][:,:,:],rsl_yt_nolvz['lon'][:],rsl_yt_nolvz['lat'][:],rsl_yt_nolvz['t'][:],lonlats[i][inds[i],0]+360,lonlats[i][inds[i],1])
```

```python
txts = ['YSV', 'SV', 'JdF']
colors = [cs.clr_n, cs.clr_c, cs.clr_s]
```

```python
fig,ax = plt.subplots(1,3,figsize=(8,2))


ax[0].plot(rsl_yt['t'][:], rsl0_yt[0],color = colors[0],label = 'Transient, LVZ')
ax[0].plot(rsl_mx['t'][:], rsl0_mx[0], color = colors[0], linestyle='dashed', label = 'Maxwell, no LVZ')
ax[0].set_xlim(22,24)
ax[0].set_ylim(0,50)
ax[0].set_xlabel('Time (ka)')
ax[0].set_ylabel('RSL (m)')
ax[0].grid()



ax[1].plot(rsl_yt['t'][:], rsl0_yt[1],color = colors[1],label = 'Transient, LVZ')
ax[1].plot(rsl_mx['t'][:], rsl0_mx[1], color = colors[1], linestyle='dashed', label = 'Maxwell, no LVZ')

ax[1].set_xlim(22,26.3)
ax[1].set_ylim(100,350)
ax[1].set_xlabel('Time (ka)')
ax[1].set_ylabel('RSL (m)')
ax[1].grid()

ax[1].legend(ncols = 2, loc = (-0.4,-0.5))



ax[2].plot(rsl_yt['t'][:], rsl0_yt[2],color = colors[2],label = 'Transient, LVZ')
ax[2].plot(rsl_mx['t'][:], rsl0_mx[2], color = colors[2], linestyle='dashed', label = 'Maxwell, no LVZ')

ax[2].set_xlim(16,18.3)
ax[2].set_ylim(0,200)
ax[2].set_xlabel('Time (ka)')
ax[2].set_ylabel('RSL (m)')
ax[2].grid()

for i in np.arange(len(colors)):
    ax[i].text(0.1,0.05,txts[i],color = colors[i], transform=ax[i].transAxes, weight = 'bold')



# plt.tight_layout()
plt.subplots_adjust(wspace=0.5)

# plt.savefig('figures/sf_maxwellVsTransient.pdf')
```

```python
legend_elements = [
                   Line2D([0], [0], color='k', lw=1, label='Transient, Low Viscosity Zone', linestyle='solid'),
                   Line2D([0], [0], color='k', lw=1, label='Maxwell, Low Viscosity Zone', linestyle='dashdot'),
                   Line2D([0], [0], color='k', lw=1, label='Transient, no Low Viscosity Zone', linestyle='dotted'),
                   Line2D([0], [0], color='k', lw=1, label='Maxwell, no Low Viscosity Zone', linestyle='--'),
                   Patch(facecolor=cs.clr_n,
                         label='Yakutat Sea Valley'),
                   Patch(facecolor=cs.clr_c,
                         label='Skeena Valley'),
                   Patch(facecolor=cs.clr_s,
                         label='Juan de Fuca'),
                   
                  ]

# Create the figure
fig, ax = plt.subplots()
ax.legend(handles=legend_elements, loc='center', ncols = 2)

plt.show()
```
